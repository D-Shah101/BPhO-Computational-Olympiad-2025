```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
def sellmeier(wavelength: np.ndarray):
    """
    Returns the crown glass refractive indexes for a given range of wavelengths using the Sellmeier formula.

    Args:
        wavelength (np.ndarray): The range of wavelength in microns.

    Returns:
        n (np.ndarray): An array of refractive indexes of crown glass for the given wavelengths.
    """
    # Sellmeier coefficients
    a = np.array([1.03961212, 0.231792344, 1.01046945])
    b = np.array([0.00600069867, 0.0200179144, 103.560653])

    wavelength_sq = np.square(wavelength)

    # Calculate Sellmeier terms using broadcasting
    terms = (a[:, np.newaxis] * wavelength_sq) / (wavelength_sq - b[:, np.newaxis])
    
    # Sum across coefficients and add 1, then take square root
    n_squared = 1 + np.sum(terms, axis=0)
    n = np.sqrt(n_squared)
    
    return n
```

```python
wavelengths = np.linspace(400, 800, 100) 
n = sellmeier(wavelengths / 1000)

data = pd.DataFrame({
    'Wavelength / nm': wavelengths,
    'Refractive index': n
})
```

```python
fig = px.line(data, x='Wavelength / nm', y='Refractive index', 
              title='Refractive Index of Crown Glass over 400-800nm')
fig.update_layout(hovermode='x unified')
fig.show()
```
