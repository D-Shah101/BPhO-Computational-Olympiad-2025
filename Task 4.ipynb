```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
c = 3e8 # m/s
n_1 = 1 # refractive index of air
n_2 = 1.5 # refractive index of glass
L = 2 # m (arbitrary value)
y = 1 # m (arbitrary value)
Y = y # (arbitrary value)
```

```python
def get_times(x: np.ndarray) -> np.ndarray:
    """
    Returns an array for the respective time taken for a light ray using Fetmat's principle for the given range of distances.

    Args:
        x (np.ndarray): Range of distances < L

    Returns:
        np.array: Range of time of flight values.
    """
    # equation: LHS = RHS_1 + RHS_2
    rhs_1 = (np.sqrt(np.square(x) + y**2)) / (c / n_1)
    rhs_2 = (np.sqrt((L - x)**2 + y**2)) / (c / n_2)
    return rhs_1 + rhs_2
```

```python
x = np.linspace(0, L, 200) # range of values 0 <= x <= L, we use 200 values for a smooth parabola
times = get_times(x)

data = pd.DataFrame({
    'x / m': x,
    'Time of flight / ns': times * 10e8
})
```

```python
fig = px.line(data, x='x / m', y='Time of flight / ns', title="Time of Flight of Light Against Glass")
fig.update_layout(hovermode='x unified')
fig.show()
```

```python
min_time_index = np.argmin(times)
min_x = x[min_time_index]
min_time_ns = times[min_time_index] * 1e9

theta = np.arctan(min_x / y)
phi = np.arctan((L - min_x) / Y)

lhs = n_1 * np.sin(theta)
rhs = n_2 * np.sin(phi)

print('--- Evaluation ---')
print(f'L = {L}m')
print(f"Calculated minimum time: {min_time_ns:.2f} ns at x = {min_x:.2f} m")
print(f"Calculated n\u2081sin\u03B8: {lhs:.3f}")
print(f"Calculated n\u2082sin\u0278: {rhs:.3f}")


if np.isclose(lhs, rhs, 0.01): # allow for small tolerance
    print("At the minimum value of x, n\u2081sin\u03B8 approximately equals n\u2082sin\u0278 which supports Snell's Law.")
else:
    print("At the minimum value of x, n\u2081sin\u03B8 deviates from n\u2082sin\u0278 which suggests discrepancies in Snell's Law or experimental errors.")
```
