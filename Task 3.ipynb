```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
c = 3e8 # m/s
n = 1.5 # refractive index of glass
L = 2 # m (arbitrary value)
y = 1 # m (arbitrary value)
```

```python
def get_times(x: np.ndarray) -> np.ndarray:
    """
    Returns an array for the respective time taken for a light ray using Fetmat's principle for the given range of distances.

    Args:
        x (np.ndarray): Range of distances < L

    Returns:
        np.array: Range of time of flight values.
    """
    # equation: LHS = RHS_1 + RHS_2
    rhs_1 = (np.sqrt(np.square(x) + y**2)) / (c / n)
    rhs_2 = (np.sqrt((L - x)**2 + y**2)) / (c / n)
    return rhs_1 + rhs_2
```

```python
x = np.linspace(0, L, 200) # range of values 0 <= x <= L, we use 200 values for a smooth parabola
times = get_times(x)

data = pd.DataFrame({
    'x / m': x,
    'Time of flight / ns': times * 10e8
})
```

```python
fig = px.line(data, x='x / m', y='Time of flight / ns', title="Time of Flight of Light Against Glass")
fig.update_layout(hovermode='x unified')
fig.show()
```

```python
min_time_index = np.argmin(times)
min_x = x[min_time_index]
min_time_ns = times[min_time_index] * 1e9

print('--- Evaluation ---')
print(f'L = {L}m')
print(f"Calculated minimum time: {min_time_ns:.2f} ns at x = {min_x:.2f} m")

if np.isclose(min_x, L / 2, 0.01): # allow for small tolerance
    print("Calculated minimum x is approximately half the length which supports Fermat's principle, implying time is minimised at x = L/2 and angle of incidence = angle of reflection.")
else:
    print("Calculated minimum x deviates from half the length which suggests discrepancies in Fermat's principle or experimental errors.")
```
