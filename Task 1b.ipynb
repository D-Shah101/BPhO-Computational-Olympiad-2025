```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
def refractive_indexes(frequencies: np.ndarray) -> np.ndarray:
    """
    Returns an array of refractive indexes of water for the given range of frequencies.

    Args:
        frequencies (np.ndarray): The range of frequences in THz.

    Returns:
        np.ndarray: An array of refractive indexes of water for the given frequencies.
    """

    freq_in_hz = frequencies * 10e12

    # LHS = (n^2-1)^-2
    # RHS = 1.731 - 0.261(f/10^15 Hz)^2
    rhs = 1.731 - 0.261*(np.square(freq_in_hz/10e15))
    lhs = np.sqrt(rhs**-1)
    n_squared = lhs + 1

    return np.sqrt(n_squared)
```

```python
def map_colour(frequency: int) -> str:
    """
    Get's the visible light colour of a given frequency.

    Args:
        frequency (int): The frequency in THz.

    Returns:
        str: The visible colour the frequency corresponds to.
    """
    if frequency < 405:
        return 'Infrared'
    if frequency >= 405 and frequency < 480:
        return 'Red'
    elif frequency >= 480 and frequency < 510:
        return 'Orange'
    elif frequency >= 510 and frequency < 530:
        return 'Yellow'
    elif frequency >= 530 and frequency < 600:
        return 'Green'
    elif frequency >= 600 and frequency < 620:
        return 'Cyan'
    elif frequency >= 620 and frequency < 680:
        return 'Blue'
    elif frequency >= 680 and frequency < 790:
        return 'Violet'
    else: # > 790
        return 'Ultraviolet'
```

```python
frequencies = np.linspace(300, 900, 100) # THz
n = refractive_indexes(frequencies)

data = pd.DataFrame({
    "Frequency / THz": frequencies,
    "Refractive index": n
})

data['Colours'] = data['Frequency / THz'].apply(map_colour)
```

```python
fig = px.line(data, x='Frequency / THz', y='Refractive index', title='Refractive Index of Water over 40',
              color='Colours', line_group='Colours')

fig.update_layout(hovermode='x unified')

colour_map = {
    'Infrared': 'rgb(128, 0, 0)', # Dark red
    'Red': 'red',
    'Orange': 'orange',
    'Yellow': 'yellow',
    'Green': 'green',
    'Cyan': 'cyan',
    'Blue': 'blue',
    'Violet': 'purple',
    'Ultraviolet': 'rgb(75, 0, 130)' # Indigo/dark purple
}

for category, colour in colour_map.items():
    fig.for_each_trace(lambda trace: trace.update(line_color=colour) if trace.name == category else None)
```

```python
fig.show()
```
